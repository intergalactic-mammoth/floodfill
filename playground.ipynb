```python
%matplotlib inline

%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
from floodfill import floodfill
import numpy as np
import PIL
import time
```

```python
#Choose the preferred resulting color
col = [100, 200, 200]

# Choose the starting point of the algorithm (root node)
start = (0, 0)

# Convert the input image to a numpy array
img = PIL.Image.open('imgs/100x100_irregular.jpg')
img = np.asarray(img).copy()

# Plot the original image
plt.imshow(img)
plt.savefig('imgs/before.png', dpi=300)
```

```python
# Perform floodfill algorithm
floodfill(img, start, col);

# Show resulting image
plt.imshow(img)
plt.savefig('imgs/after.png', dpi=300)
```

```python
#Try with new root and other color
new_start = (60, 60)
col_new = [50, 125, 250]
floodfill(img, new_start, col_new)

plt.imshow(img)
```

```python
# Measurement to check if time to solution is increasing
# linearly with the number of pixels
imgs = []
pixels = []

# 100 pixels
img = PIL.Image.open('imgs/10x10.jpg')
img = np.asarray(img).copy()
imgs.append(img)
pixels.append(img.size)

# 200 pixels
img = PIL.Image.open('imgs/20x10.jpg')
img = np.asarray(img).copy()
imgs.append(img)
pixels.append(img.size)

# 400 pixels
img = PIL.Image.open('imgs/20x20.jpg')
img = np.asarray(img).copy()
imgs.append(img)
pixels.append(img.size)

# 800 pixels
img = PIL.Image.open('imgs/40x20.jpg')
img = np.asarray(img).copy()
imgs.append(img)
pixels.append(img.size)

#Run for all 4 images and print times
for i in range(len(imgs)):
    begin = time.time()
    floodfill(imgs[i], start, col);
    end = time.time()
    print("%.2f seconds for %d pixels" % ((end-begin), pixels[i]))

# We see that indeed the time to solution is increasing linearly.
```
