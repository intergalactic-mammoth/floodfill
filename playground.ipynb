```python
%matplotlib inline

%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
from floodfill import floodfill
import numpy as np
import PIL
import time
```

```python
#Choose the preferred resulting color
col = [200, 100, 20]

# Choose the starting point of the algorithm (root node)
u = 0; v = 0

# Convert the input image to a numpy array
img = PIL.Image.open('test_harder.jpg')
img = np.asarray(img).copy()

# Plot the original image
plt.imshow(img)
```

```python
# Perform floodfill algorithm
floodfill(img, u, v, col);

# Show resulting image
plt.imshow(img)
```

```python
# Measurement to check if time to solution is increasing
# linearly with the number of pixels
imgs = []

# 100 pixels
img = PIL.Image.open('10x10.jpg')
img = np.asarray(img).copy()
imgs.append(img)

# 200 pixels
img = PIL.Image.open('20x10.jpg')
img = np.asarray(img).copy()
imgs.append(img)

# 400 pixels
img = PIL.Image.open('20x20.jpg')
img = np.asarray(img).copy()
imgs.append(img)

# 800 pixels
img = PIL.Image.open('40x20.jpg')
img = np.asarray(img).copy()
imgs.append(img)

#Run for all 4 images and print times
for img in imgs:
    begin = time.time()
    floodfill(img, u, v, col);
    end = time.time()
    print("Took %.2f seconds" % (end-begin))

# We see that indeed the time to solution is increasing linearly.
```
